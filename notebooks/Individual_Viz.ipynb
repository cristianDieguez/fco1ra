```python
def draw_player_card(
    img_path,
    name,
    dorsal,
    position,
    total_minutes,
    avg_minutes,
    goals,
    assists,
    figsize=(10,4)
):
    fig, ax = plt.subplots(figsize=figsize)
    ax.set_facecolor("#111111")
    fig.patch.set_facecolor("#111111")
    ax.axis("off")

    # Load local image
    try:
        player_img = Image.open(img_path)
        zoom = 3
        w, h = player_img.size
        player_img = player_img.resize((int(w * zoom), int(h * zoom)))


    except:
        player_img = Image.new("RGB", (300,300), color=(40,40,40))

    player_img = np.array(player_img)

    # Left image section
    ax_img = fig.add_axes([0.05, 0.1, 0.28, 0.8])
    ax_img.imshow(player_img)
    ax_img.axis("off")

    # Right stats card
    ax_stats = fig.add_axes([0.37, 0.1, 0.58, 0.8])
    ax_stats.axis("off")

    bg = FancyBboxPatch(
        (0, 0), 1, 1,
        boxstyle="round,pad=0.02,rounding_size=0.03",
        linewidth=1,
        edgecolor="#2A2A2A",
        facecolor="#1A1A1A"
    )
    ax_stats.add_patch(bg)

    y = 0.85
    gap = 0.14

    # Name & Dorsal
    ax_stats.text(0.05, y, name, fontsize=20, color="white", weight="bold",
                  ha="left", va="center", transform=ax_stats.transAxes)
    ax_stats.text(0.60, y, f"#{dorsal}", fontsize=20, color="white",
                  ha="left", va="center", transform=ax_stats.transAxes)
    y -= gap

    # Position
    ax_stats.text(0.05, y + 0.05, position, fontsize=14, color="#BBBBBB",
                  ha="left", va="center", transform=ax_stats.transAxes)
    y -= gap * 1.1

    # Total Minutes
    ax_stats.text(0.05, y, "Total Minutos", fontsize=14, color="#888888",
                  ha="left", transform=ax_stats.transAxes)
    ax_stats.text(0.60, y, f"{format_mmss(total_minutes)}", fontsize=16, color="white",
                  ha="left", transform=ax_stats.transAxes)
    y -= gap

    # Average Minutes
    ax_stats.text(0.05, y, "Promedio / Partido", fontsize=14, color="#888888",
                  ha="left", transform=ax_stats.transAxes)
    ax_stats.text(0.60, y, f"{format_mmss(avg_minutes)}", fontsize=16, color="white",
                  ha="left", transform=ax_stats.transAxes)
    y -= gap * 1.2

    # Goals
    ax_stats.text(0.05, y, "Goles", fontsize=14, color="#888888",
                  ha="left", transform=ax_stats.transAxes)
    ax_stats.text(0.60, y, f"{goals}", fontsize=16, color="white",
                  ha="left", transform=ax_stats.transAxes)
    y -= gap

    # Assists
    ax_stats.text(0.05, y, "Asistencias", fontsize=14, color="#888888",
                  ha="left", transform=ax_stats.transAxes)
    ax_stats.text(0.60, y, f"{assists}", fontsize=16, color="white",
                  ha="left", transform=ax_stats.transAxes)

    return fig
```
